```python
import pandas as pd
import re
```

```python
f = open('analysis.csv', 'r')
lines = f.readlines()
output = []
for line in lines:
    
    if "[" not in line:
        output.append(line)
    else:
        
        new_line = ""
        
        mode = 0
        for c in line:
            if c != "[" and c != "]" and mode == 0:
                new_line += c
            elif c == "[" and mode == 0:
                new_line += c
                mode = 1
            elif c == "]" and mode == 1:
                new_line += c
                mode = 0
            elif c == "," and mode == 1:
                new_line += ";"
            else:
                new_line += c
        print(new_line)
        
        output.append(new_line)
    
f = open('analysis_fixed.csv', 'w')
for line in output:
    f.write(line + "\n")
        
f.close()
        
    
```
