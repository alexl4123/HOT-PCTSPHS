```python
import csv
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import math
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
df_m_init = pd.read_csv("m_init.csv")
```

```python
df_m_init
```

```python
data = df_m_init
```

```python
iterations = list(data["iteration"])
iterations.sort() # Just to be safe
iterations = set(iterations)
```

```python
score_first = 0
score_second = 0

avg_score_first = 0
avg_score_second = 0

for iteration in iterations:
    tmp_df = data[data["iteration"] == int(iteration)]
    
    first = tmp_df.iloc[0]
    second = tmp_df.iloc[1]
    
    
    if float(first["delta"]) == 0.5:
        first_option = first
        second_option = second
    else:
        first_option = second
        second_option = first
        
        
        
    first_option_str = first_option["fitness"][1:len(list(first_option["fitness"])) - 1]
    first_fitness_list = list(map(float, first_option_str.split(';')))
    first_fitness_list.sort()
    second_option_str = second_option["fitness"][1:len(list(second_option["fitness"])) - 1]
    second_fitness_list = list(map(float, second_option_str.split(';')))
    second_fitness_list.sort()
    
    
    
    for index in range(len(first_fitness_list)):
        if first_fitness_list[index] > second_fitness_list[index]:
            score_first += 1
        else:
            score_second += 1
            
    avg_first = sum(first_fitness_list) / len(first_fitness_list)
    avg_second = sum(second_fitness_list) / len(second_fitness_list)
    
    if avg_first > avg_second:
        avg_score_first += 1
    else:
        avg_score_second += 1

    

    
    
    
    #size_per_iteration.append(tmp_df.shape[0])
    #log_size_per_iteration.append(math.log(tmp_df.shape[0]))
```

```python
print(score_first)
print(score_second)
```
