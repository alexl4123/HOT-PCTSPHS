```python
data = df_m_init
```

```python
iterations = list(data["iteration"])
iterations.sort() # Just to be safe
iterations = set(iterations)
```

```python
average_scores = [0] * 4

for iteration in iterations:
    tmp_df = data[data["iteration"] == int(iteration)]
        
    fitness_lists = []
        
    for i in range(4):
        fitness_str = (tmp_df.iloc[i])["fitness"][1:len(list((tmp_df.iloc[i])["fitness"])) - 1]
        fitness_list = list(map(float, fitness_str.split(';')))
        fitness_list.sort()          
        
        fitness_lists.append(fitness_list)
        
    averages = []
    for fitness_list in fitness_lists:            
         averages.append(sum(fitness_list) / len(fitness_list))
            
    best_index = averages.index(max(averages))
    average_scores[best_index] += 1
    
    
    """
    if avg_first > avg_second:
        avg_score_first += 1
    else:
        avg_score_second += 1
    """
    

    
    
    
    #size_per_iteration.append(tmp_df.shape[0])
    #log_size_per_iteration.append(math.log(tmp_df.shape[0]))
```
